$$
\frac{d\sigma}{d\Omega} = |f(\theta)|^2 =\frac{1}{k^2}\sum_{l,l'}e^{i(\delta_l-\delta_{l'})}(2l+1)\sin{\delta_{l}}(2l'+1)\sin{\delta_{l'}}P_l(\cos{\theta})P_l'(\cos{\theta})\\
\sigma = \sum_{l=0}^{\infty}\sigma_l\\
\sigma_l = \frac{4\pi}{k^2}(2l+1)\sin^2{\delta_l}
$$

```python
import numpy as np
from scipy import special
from matplotlib import pyplot as plt

R = 1417
x = np.linspace(0,6/R,100)
def delta_l(l):
    jv = special.spherical_jn(l,R*x)
    yv = special.spherical_yn(l,R*x)
    return np.arctan2(jv,yv) -np.pi

for i in range(6):
    jv = special.spherical_jn(i,R*x)
    yv = special.spherical_yn(i,R*x)
    plt.plot(R*x,delta_l(i)/np.pi, label='l={}'.format(i))
    plt.xlabel(r'$kR$')
    plt.ylabel(r'$\delta_l/\pi$')
    plt.legend()
```

```python
def sigma_l(l):
    return (4*np.pi/(k**2))*(2*l+1)*(np.sin(delta_l(l)))**2

def sigma(n):
    sum = 0
    for i in range(n+1):
        sum += sigma_l(i)
    
    return sum

k = np.linspace(0,6/R,100)
n = 5
for i in range(n+1):
    plt.plot(k*R,sigma_l(i), label=r'l={}'.format(i))
    plt.xlabel(r'$kR$')
    plt.ylabel(r'$\sigma_l(\times10^7)$')
    plt.legend()
```
